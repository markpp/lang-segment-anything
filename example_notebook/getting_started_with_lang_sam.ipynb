```python
# ~/ml/bin/python3

import warnings
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
from lang_sam import LangSAM

image = '/home/markpp/GitHub/pprobotics/Lego/frames/IMG_1936.jpeg'
# image = "./image.jpg"
text_prompt = "lego bricks"

def download_image(url):
    response = requests.get(url)
    response.raise_for_status()
    return Image.open(BytesIO(response.content)).convert("RGB")

def save_mask(mask_np, filename):
    mask_image = Image.fromarray((mask_np * 255).astype(np.uint8))
    mask_image.save(filename)

def display_image_with_masks(image, masks, save=False):
    num_masks = len(masks)

    fig, axes = plt.subplots(1, num_masks + 1, figsize=(15, 5))
    axes[0].imshow(image)
    axes[0].set_title("Original Image")
    axes[0].axis('off')

    for i, mask_np in enumerate(masks):
        axes[i+1].imshow(mask_np, cmap='gray')
        axes[i+1].set_title(f"Mask {i+1}")
        axes[i+1].axis('off')

    plt.tight_layout()
    if save:
        plt.savefig("image_with_masks.png")
    plt.show()

def display_image_with_boxes(image, boxes, logits, save=False):
    fig, ax = plt.subplots()
    ax.imshow(image)
    ax.set_title("Image with Bounding Boxes")
    ax.axis('off')

    for box, logit in zip(boxes, logits):
        x_min, y_min, x_max, y_max = box
        confidence_score = round(logit.item(), 2)  # Convert logit to a scalar before rounding
        box_width = x_max - x_min
        box_height = y_max - y_min

        # Draw bounding box
        rect = plt.Rectangle((x_min, y_min), box_width, box_height, fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)

        # Add confidence score as text
        ax.text(x_min, y_min, f"Confidence: {confidence_score}", fontsize=8, color='red', verticalalignment='top')

    if save:
        plt.savefig("image_with_boxes.png")
    plt.show()

def print_bounding_boxes(boxes):
    print("Bounding Boxes:")
    for i, box in enumerate(boxes):
        print(f"Box {i+1}: {box}")

def print_detected_phrases(phrases):
    print("\nDetected Phrases:")
    for i, phrase in enumerate(phrases):
        print(f"Phrase {i+1}: {phrase}")

def print_logits(logits):
    print("\nConfidence:")
    for i, logit in enumerate(logits):
        print(f"Logit {i+1}: {logit}")

def main():
    # Suppress warning messages
    warnings.filterwarnings("ignore")

    try:
        if image.startswith("http"):
            image_pil = download_image(image)
        else:
            image_pil = Image.open(image).convert("RGB")

        model = LangSAM()
        masks, boxes, phrases, logits = model.predict(image_pil, text_prompt)

        if len(masks) == 0:
            print(f"No objects of the '{text_prompt}' prompt detected in the image.")
        else:
            # Convert masks to numpy arrays
            masks_np = [mask.squeeze().cpu().numpy() for mask in masks]

            # Display the original image and masks side by side
            display_image_with_masks(image_pil, masks_np, save=True)

            # Display the image with bounding boxes and confidence scores
            display_image_with_boxes(image_pil, boxes, logits, save=True)

            # Save the masks
            for i, mask_np in enumerate(masks_np):
                mask_path = f"image_mask_{i+1}.png"
                save_mask(mask_np, mask_path)

            # Print the bounding boxes, phrases, and logits
            print_bounding_boxes(boxes)
            print_detected_phrases(phrases)
            print_logits(logits)

    except (requests.exceptions.RequestException, IOError) as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
import requests
import sys, os
from PIL import Image
import cv2
from io import BytesIO
from lang_sam import LangSAM

video = '/home/markpp/Datasets/LEGO/slopes_arcs_rainbows/Slopes_arcs_and_rainbows-08.mp4'
text_prompt = "lego brick"


def print_bounding_boxes(boxes):
    print("Bounding Boxes:")
    for i, box in enumerate(boxes):
        print(f"Box {i+1}: {box}")

def print_detected_phrases(phrases):
    print("\nDetected Phrases:")
    for i, phrase in enumerate(phrases):
        print(f"Phrase {i+1}: {phrase}")

def print_logits(logits):
    print("\nConfidence:")
    for i, logit in enumerate(logits):
        print(f"Logit {i+1}: {logit}")

def main():
    # Suppress warning messages
    warnings.filterwarnings("ignore")

    # make sure the output directory exists
    if not os.path.exists('detections'):
        os.makedirs('detections')

    # make sure the output directory exists
    if not os.path.exists('crops'):
        os.makedirs('crops')

    model = LangSAM()
    
    cap = cv2.VideoCapture(video)
    if not cap.isOpened():
        print("Error: Could not open video.")
        sys.exit()

    frame_number = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        if frame_number % 10 == 0:
            # Convert image to RGB if it's in BGR format
            if frame.shape[-1] == 3:
                rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            masks, boxes, phrases, logits = model.predict(Image.fromarray(rgb_frame), text_prompt)

            if len(masks) == 0:
                print(f"No objects of the '{text_prompt}' prompt detected in the image.")
            else:
                # Convert masks to numpy arrays
                masks_np = [(mask.squeeze().cpu().numpy()*255).astype(np.uint8) for mask in masks]

                for box_id, (box, logit, mask) in enumerate(zip(boxes, logits, masks_np)):
                    x_min, y_min, x_max, y_max = map(int, box)
                    confidence_score = round(logit.item(), 2)  # Convert logit to a scalar before rounding

                    if confidence_score > 0.4:
                        crop_path = f'crops/frame_{frame_number}_crop_{box_id}.jpg'
                        cv2.imwrite(crop_path, frame[y_min:y_max, x_min:x_max])
                        cv2.imwrite(crop_path.replace('.jpg','.png'), mask[y_min:y_max, x_min:x_max])

                        # Draw bounding box
                        cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 0, 255), 2)

                        # Add confidence score as text
                        cv2.putText(frame, f"Confidence: {confidence_score}", (x_min, y_min - 10), 
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2, cv2.LINE_AA)
                
                cv2.imwrite(f'detections/frame_{frame_number}.png', frame)

                # Print the bounding boxes, phrases, and logits
                print_bounding_boxes(boxes)
                #print_detected_phrases(phrases)
                print_logits(logits)

        frame_number += 1

if __name__ == "__main__":
    main()
```
